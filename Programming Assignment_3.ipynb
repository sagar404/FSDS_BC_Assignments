```python
#1. Write a Python Program to Check if a Number is Positive, Negative or Zero?


def check_integer(num):
    '''
    This function will check if given number is Positive, Negative or Zero

    num -> user input number 

    '''
    try:
        #validating num is an integer
        if type(num) != int:
            raise Exception("Error: provided input is not an integer")
        #checking num is Positive, Negative or Zero
        elif num>0:
            return f'{num} is a Positive number'
        elif num<0:
            return f'{num} is a Negative number'
        elif num == 0:
            return f'Hi, I am Zero'
    except Exception as e:
        return e 

```

```python
check_integer(5)
```

```python
#2. Write a Python Program to Check if a Number is Odd or Even?

def check_odd_even(num):

    '''
    This is a function to check if a Number is Odd or Even
    num -> user input number 

    '''
    try :
        #validating num is an integer
        if type(num) != int:
            raise Exception("Error: provided input is not an integer")
        #checking if num is Odd or Even 
        elif num % 2 == 0:
            return f'{num} is an Even number'
        else:
            return f'{num} is a Odd number'
    except Exception as e:
        return e 
```

```python
check_odd_even(4)
```

```python
#3. Write a Python Program to Check Leap Year?

def leap_year(yyyy):
    '''
    This function will check leap year 

    yyyy - > Year in yyyy format (user input) 
    '''
    try:
        #Validating wether the year provided is an integer and in yyyy format  
        if len(str(yyyy)) != 4 and type(yyyy) != int:
            raise Exception("Error: Input not in 'yyyy' format")
        elif (yyyy % 400 == 0) and (yyyy % 100 == 0):
            print("{0} is a leap year".format(yyyy))
        elif (yyyy % 4 ==0) and (yyyy % 100 != 0):
            print("{0} is a leap year".format(yyyy))
        else:
            print("{0} is not a leap year".format(yyyy))
    except Exception as e:
        return e       
```

```python
leap_year(2000)
```

```python
#4. Write a Python Program to Check Prime Number?

def check_Prime(num): 
    ''' 
    Fucttion to check a number

    num -> User input for which function will check is prime or not
    '''
    #check if number greater than 1 
    if num >1:
        for i in range(2, num):
            #cheking if number is divsible by any other number 
            if num % i == 0:
                return f'{num} is a not a Prime Number'
                break
        else:
            return f'{num} is a Prime Number'
    else:
        return f'{num} is not a Prime Number'
    
```

```python
check_Prime(10)
```
